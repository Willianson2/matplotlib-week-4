```python
import matplotlib
import matplotlib.pyplot as a
#variabel
```

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
#buat kategori dan data
```

```python
a.pie(data,
      labels=kategori,
      autopct='%1.1f%%',
      startangle=90)
a.title('contooh pie plot')
a.show()
#perintah buat grafik bulet dengan data diatas
```

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']
#buat kategori dan data dan warna
```

```python
a.pie(data,
      labels=kategori,
      colors=warna,
      autopct='%1.1f%%',
      startangle=90)
a.legend()
a.title('contoh colored pie plot')
a.show()
#buat grafik bulet dengan data diatas
```

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']
explode_var = [0., 0., 0.2, 0., 0.08]
#buat kategori dan data dan warna
```

```python
a.pie(data,
      labels=kategori,
      colors=warna,
      explode=explode_var,
      autopct='%1.1f%%',
      startangle=90)
a.title('contoh exploded pie plot')
a.show()
#membuat grafik bulet yang dipecah berdasarkan data diatas
```
