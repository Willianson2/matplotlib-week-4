```python
import matplotlib
import matplotlib.pyplot as a
import numpy as b
#variabel
```

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
#membuat data dan kategori
```

```python
a.bar(kategori, data)
a.grid()
a.xlabel('kategori')
a.ylabel('jumlah')
a.title ('contoh bar plot')
a.show()
#membuat grafik jenis bar, data 
```

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
#membuat data dan kategori baru
```

```python
a.bar(kategori,
      data,
      color='red',
      alpha=0.25) #ketebalan warna
a.grid(linestyle='--',
       linewidth=2,
       axis='x',
       alpha=0.50) #ketebalan garis putus putus
a.xlabel('kategori')
a.ylabel('jumlah')
a.title('contoh bar plot')
a.show()
#membuat grafik bar warna merah, grid garis garis putus dari garis x
```

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']
#buat data baru 2 biji dan kategori baru
```

```python
a.bar(kategori, data1, label='Data 1')
#a.bar(kategori, data2, label='Data 2')
a.bar(kategori, data2, label='Data 2', bottom=data1)
a.legend()
a.grid(linestyle='--',
       linewidth=1,
       axis='y',
       alpha=0.75)
a.xlabel('kategori')
a.ylabel('kategori')
a.title('contoh stacked bar plot')
a.show()
#membuat stacked bar plot, data1 dan data2 digabung. data 1 ditambah dan digabung data 2
```

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']
x=b.arange(len(kategori))
x
#buat data
```

```python
width = 0.35
a.bar(x-width/2, data1, width, label='Data1')
a.bar(x+width/2, data2, width, label='Data2')
a.xticks(x, kategori)
a.grid(linestyle='--',
       linewidth=1,
       axis='y',
       alpha=0.75)
a.xlabel('kategori')
a.ylabel('jumlah')
a.title('contoh grouped abr plot')
a.legend()
a.show()
#perintah untuk membuat grouped bar plot
```
